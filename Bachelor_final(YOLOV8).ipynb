```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Define function to draw text on image
def draw_text(image, text, position):
    draw = ImageDraw.Draw(image)
    font = ImageFont.load_default()
    draw.text(position, text, (255, 0, 0), font=font)

# Set the path to your output video file
output_video_path = '/content/runs/detect/predict5/IMG_6882.mp4'

# Initialize variables
vehicle_count = 0

# Open the output video
cap = cv2.VideoCapture(output_video_path)

frame_with_text_np = np.array(frame_with_text)

while True:
    ret, frame = cap.read()

    if not ret:
        break

    # Convert frame to PIL Image
    pil_frame = Image.fromarray(frame)

    # Draw vehicle count on the frame
    draw_text(pil_frame, f"Vehicle Count: {vehicle_count}", (10, 30))

    # Convert PIL Image back to numpy array
    frame_with_text = pil_frame.convert('RGB')


    # Display the frame with vehicle count
    cv2_imshow(frame_with_text)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
